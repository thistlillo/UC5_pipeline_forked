```python
import pyeddl.eddl as eddl
import pyecvl.ecvl as ecvl
from pyeddl.tensor import Tensor

import pandas as pd
import numpy as np
from posixpath import join
import yaml
import os
from utils.data_partitioning import load_data_split

# def combine_ds(ecvl_ds, uc)

exp_fld = "/opt/uc5/results/eddl_exp/wp6"
img_fld = "/mnt/datasets/uc5/std-dataset/image"

train_ids, valid_ids, test_ids = load_data_split(exp_fld)
img_ds_fn = join(exp_fld, "img_reports_phi2_enc.tsv")

ds = pd.read_csv(img_ds_fn, sep="\t")
ds.drop(columns=["Unnamed: 0"], inplace=True)
print(ds.columns)

```

```python
def ecvl_yaml(filenames, labels, train_ids, valid_ids, test_ids):
    d = {
        "name"        : "frontal-lateral",
        "description" : "iu frontal-lateral images",
        "classes"     : [], 
        "images"      : [],
        "split"       : dict(training = train_ids, 
                            validation = valid_ids, 
                            test=test_ids)
    }
    imgs = []
    for fn, l in zip(filenames, labels):
        imgs.append({
            "location": fn,
            "label": [ll for ll in l]
        })
    d["images"] = imgs
    all_labels = [lab for labs in labels for lab in labs]
    all_labels = set(all_labels)
    print(all_labels)

    d["classes"] = sorted(list(all_labels))
    return d
```

```python
# training = ds.loc[train_ids]
# assert len(train_ids) == training.shape[0]

ds["path"] = ds.filename.apply(lambda x: join(img_fld, x))
ds["labels_list"] = ds.labels.apply(lambda x: list(set(int(s) for s in x.split(";"))))
ds = ds.set_index("filename")
# ds["path"] = ds.reset_index()["filename"]
display(ds)
```

```python
filenames = []
labels = []
counter = 0
for ids in [train_ids, valid_ids, test_ids]:
    split_filenames = ds.loc[ids, "path"].tolist()
    filenames = filenames + split_filenames
    assert len(split_filenames) == len(ids)
    split_labels = ds.loc[ids, "labels_list"].tolist()
    labels = labels + split_labels
    assert len(split_labels) == len(ids)

print(len(filenames))
print(len(labels))

n_training = len(train_ids)
n_valid = len(valid_ids)
n_test = len(test_ids)

train_r = list(range(0, n_training))
valid_r = list(range(n_training, n_training + n_valid))
test_r = list(range(n_test, n_test+ n_training + n_valid))

y = ecvl_yaml(filenames, labels, train_r, valid_r, test_r)
out_fn = join(exp_fld, "dataset.yml")
with open(out_fn, "w") as fout:
    yaml.safe_dump(y, fout, default_flow_style=True)
print("saved:", out_fn)
```
